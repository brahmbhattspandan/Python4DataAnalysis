```python
# Create a DF containing data. Index are 2000-2015, columns are ['College', 'Admits', 'GPA']
import numpy as np
narr = np.arange(48).reshape((16,3))
ind = np.arange(2000,2016)
sdf = pd.DataFrame(narr, index=ind, columns = ['College', 'Admits', 'GPA'])
sdf
```

```python
# Add a column called as fees and start fees by 1000 and increment it by 100 every year.
```

```python
end = 1000 + 100* len(sdf)
f = np.arange(1000, end,100)
sdf['fees'] = f
sdf
```

```python
# Create new DF by droping data before 2005
ind = sdf[sdf.index < 2005].index
nsdf = sdf.drop(ind)
nsdf
```
